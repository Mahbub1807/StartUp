```python
from tkinter import*
import math
```

```python
def close_window():
    expression = "" 
    root.destroy()
    
```

```python
def move_calculator(e):
    root.geometry(f'+{e.x_root}+{e.y_root}')
```

```python
# globally declare the expression variable 
expression = "" 

def sqRoot():
    try:
        global expression
        x = float(expression)
        sqrt = math.sqrt(x)
        expression = str(sqrt)
        display.config(text = expression)
    except:
        pass
    
def disable_btn():
    button26["state"] = DISABLED

def press(num): 
    global expression 
 
    # concatenation of string 
    expression = expression + str(num) 
 
    # update the expression by using set method 
    display.config(text = expression) 
 
 
# Function to evaluate the final expression 
def equalpress(): 
    try: 
        global expression
        expression = expression.replace("x","*")
        expression = expression.replace("÷","/")
        total = str(eval(expression)) 
 
        display.config(text = total) 

        expression = total 

    except:
        expression = "" 

def clear(): 
    global expression 
    expression = "" 
    display.config(text = "0")
    
def clearprev():
    global expression
    length = len(expression)
    if (length==1) | (length==0):
        expression = ""
        display.config(text = "0")
    else:
        expression = expression[0:length-1]
        display.config(text = expression)
        
def signChange():
    global expression
    x = float(expression)
    x = x* -1
    expression = str(x)
    display.config(text = expression)
      
```

```python
root = Tk()
root.geometry("270x420+300+300")
root.configure(background="#d3d3d3")

root.overrideredirect(True)

cal_window = Canvas(root, bg="#d3d3d3",bd=5)
cal_window.pack(fill="both", expand=True)

close_label = Frame(cal_window,bg="#d3d3d3")
close_label.pack(fill="both",expand=True)
close_label.bind("<B1-Motion>",move_calculator)
close_button_container = Label(close_label,bg="#000000",bd=0)
close_button_container.pack(side=RIGHT,padx=10,pady=4)
close_button = Button(close_button_container, text="X",bg="#d3d3d3",height=1,width=2,font=("Times, 11"),
                      command = close_window,highlightthickness = 2,bd=0)
close_button.pack(padx=1,pady=1)

window_dis=Frame(cal_window,bg="#d3d3d3")
window_dis.pack(fill="both",expand=True)
display = Label(window_dis, text="0", height=2, width=26, font =("Times, 13"), bg="#ffffff",anchor="e",justify="right")
display.pack(fill="both",padx=10,pady=5)

window = Frame(cal_window, bg="#d3d3d3")
window.pack(fill="both",expand=True)

label0 = Frame(window, bg = "#d3d3d3")
label0.grid(row=1,column=0,padx=5)

button1 = Button(window, text='MR', fg='white', bg="#234761", height=2, width=4,font =("Times", 11,"bold"),) 
button1.grid(row=2, column=1,pady=2,sticky = 'N')

button2 = Button(window, text='MC', fg='white', bg="#234761", height=2, width=4,font =("Times", 11,"bold"),) 
button2.grid(row=2, column=2,pady=2,sticky = 'N')

button3 = Button(window, text='M+', fg='white', bg="#234761", height=2, width=4,font =("Times", 11,"bold"),) 
button3.grid(row=2, column=3,pady=2,sticky = 'N')

button4 = Button(window, text='(', fg='white', bg="#4E7991",command=lambda: press("("), height=2, width=4,font =("Times", 11,"bold"),) 
button4.grid(row=2, column=4,pady=2,sticky = 'N')

button5 = Button(window, text=')', fg='white', bg="#9cb3c1",command=lambda: press(")"), height=2, width=4,font =("Times", 11,"bold"),) 
button5.grid(row=2, column=5,pady=2,sticky = 'N')

button6 = Button(window, text='7', fg='black', bg="#ffffff",command=lambda: press(7), height=2, width=4,font =("Times", 11,"bold"),) 
button6.grid(row=3, column=1,pady=2,sticky = 'N')

button7 = Button(window, text='8', fg='black', bg="#ffffff",command=lambda: press(8), height=2, width=4,font =("Times", 11,"bold"),) 
button7.grid(row=3, column=2,pady=2,sticky = 'N')

button8 = Button(window, text='9', fg='black', bg="#ffffff",command=lambda: press(9), height=2, width=4,font =("Times", 11,"bold"),) 
button8.grid(row=3, column=3,pady=2,sticky = 'N')

button9 = Button(window, text='÷', fg='white', bg="#4E7991",command=lambda: press("÷"), height=2, width=4,font =("Times", 11,"bold"),) 
button9.grid(row=3, column=4,pady=2,sticky = 'N')

button10 = Button(window, text='C', fg='white', bg="#BA5B23",command=lambda: clear(), height=2, width=4,font =("Times", 11,"bold"),) 
button10.grid(row=3, column=5,pady=2,sticky = 'N')

button11 = Button(window, text='4', fg='black', bg="#ffffff", command=lambda: press(4),height=2, width=4,font =("Times", 11,"bold"),) 
button11.grid(row=4, column=1,pady=2,sticky = 'N')

button12 = Button(window, text='5', fg='black', bg="#ffffff",command=lambda: press(5), height=2, width=4,font =("Times", 11,"bold"),) 
button12.grid(row=4, column=2,pady=2,sticky = 'N')

button13 = Button(window, text='6', fg='black', bg="#ffffff",command=lambda: press(6), height=2, width=4,font =("Times", 11,"bold"),) 
button13.grid(row=4, column=3,pady=2,sticky = 'N')

button14 = Button(window, text='x', fg='white', bg="#4E7991",command=lambda: press("x"), height=2, width=4,font =("Times", 11,"bold"),) 
button14.grid(row=4, column=4,pady=2,sticky = 'N')

button15 = Button(window, text='CE', fg='white', bg="#BA5B23",command=lambda: clearprev(), height=2, width=4,font =("Times", 11,"bold"),) 
button15.grid(row=4, column=5,pady=2,sticky = 'N')

button16 = Button(window, text='1', fg='black', bg="#ffffff", command=lambda: press(1),height=2, width=4,font =("Times", 11,"bold"),) 
button16.grid(row=5, column=1,pady=2,sticky = 'N')

button17 = Button(window, text='2', fg='black', bg="#ffffff",command=lambda: press(2), height=2, width=4,font =("Times", 11,"bold"),) 
button17.grid(row=5, column=2,pady=2,sticky = 'N')

button18 = Button(window, text='3', fg='black', bg="#ffffff",command=lambda: press(3), height=2, width=4,font =("Times", 11,"bold"),) 
button18.grid(row=5, column=3,pady=2,sticky = 'N')

button19 = Button(window, text='-', fg='white', bg="#4E7991",command=lambda: press("-"), height=2, width=4,font =("Times", 11,"bold"),) 
button19.grid(row=5, column=4,pady=2,sticky = 'N')

button20 = Button(window, text='√', fg='white', bg="#4E7991",command=lambda: sqRoot(), height=2, width=4,font =("Times", 11,"bold"),) 
button20.grid(row=5, column=5,pady=2,sticky = 'N')

button21 = Button(window, text='±', fg='white', bg="#4E7991",command=lambda: signChange(), height=2, width=4,font =("Times", 11,"bold"),) 
button21.grid(row=6, column=1,pady=2,sticky = 'N')

button22 = Button(window, text='0', fg='black', bg="#ffffff",command=lambda: press(0), height=2, width=4,font =("Times", 11,"bold"),) 
button22.grid(row=6, column=2,pady=2,sticky = 'N')

button23 = Button(window, text='.', fg='black', bg="#ffffff",command=lambda: press("."), height=2, width=4,font =("Times", 11,"bold"),) 
button23.grid(row=6, column=3,pady=2,sticky = 'N')

button24 = Button(window, text='+', fg='white', bg="#4E7991",command=lambda: press("+"), height=2, width=4,font =("Times", 11,"bold"),) 
button24.grid(row=6, column=4,pady=2,sticky = 'N')

button25 = Button(window, text='=', fg='white', bg="#234761",command=lambda: equalpress(), height=2, width=4,font =("Times", 11,"bold"),) 
button25.grid(row=6, column=5,pady=2,sticky = 'N')

button26 = Button(window, text='Transfer Display', fg='white', bg="#A2A8AB",
                  command=disable_btn,height=2,width=34,font =("Times", 10,"bold"),) 
button26.grid(row=7, column=1,columnspan=5,pady=2,sticky = 'N')








root.mainloop()
```
